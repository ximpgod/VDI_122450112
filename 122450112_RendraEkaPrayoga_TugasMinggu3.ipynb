```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/diabetes.csv')
correlation_matrix = df.corr() #menghitung korelasi pearson dari semua target
target_correlation = correlation_matrix[['Outcome']].sort_values(by='Outcome', ascending=False) #ambil korelasi dari fitur target 'Outcome' dan mengurutkannya dari yang paling tinggi pengaruhnya

#visualisasi heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(target_correlation, annot=True, cmap='YlGn', linewidths=0.5)
plt.title('Korelasi Pearson dengan Atribut Outcome', fontsize=16)

plt.show()
```

Penjelasan :



*   Visualisasi di atas digunakan untuk melihat kemungkinan terjadinya penyakit diabetes berdasarkan berbagai kemungkinan berupa fitur-fitur seperti umur, BMI, kehamilan, tingkat glukosa, dll. Fitur 'Outcome' digunakan sebagai fitur target dimana merupakan fitur boolean yang menentukan apakah seseorang memiliki penyakit diabetes atau tidak berdasarkan data fitur lainnya
*   Apabila nilai mendekati 1, maka korelasinya akan semakin tinggi, yang mengartikan bahwa fitur tersebut berpengaruh besar terhadap terjadinya penyakit diabetes pada seseorang, dan juga sebaliknya.




Analisis Hasil Visualisasi

*   Tujuan : Memberikan informasi kepada pembaca mengenai faktor-faktor yang berpengaruh besar terhadap kemungkinan penyakit diabetes pada seseorang

*   User : Sasaran utama dari visualisasi ini adalah orang-orang yang peduli dan awas mengenai adanya kemungkinan penyakit diabetes pada diri mereka, sehingga bisa melakukan tindakan pencegahan. Hal ini berguna karena visualisasi tersebut mengurutkan korelasi beberapa faktor penyebab dengan kemungkinan terjangkitnya seseorang

*   Fungsi : Mengingatkan pembaca bahwa tingkat glukosa dalam tubuh menjadi faktor terbesar pertanda diabetes dalam tubuh.

*   Tone : Visualisasi menggunakan gradasi warna hijau yang mudah untuk dicerna oleh orang awam, semakin tua warna hijau maka akan semakin tinggi korelasi fitur terhadap fitur target, dan semakin pucat maka kemungkinan kecil fitur tersebut berpengaruh terhadap fitur target.
